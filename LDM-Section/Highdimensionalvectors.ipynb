```python
import numpy as np

```

```python

# Step 1: Define the multivariate Gaussian
dim = 500  # Dimension
mu = np.ones(dim)  # Mean vector
Sigma = np.diag([0.1] * dim)  # Diagonal covariance matrix

n=100  # Number of vectors drawn


# Initialize lists to store the results
least_squares_weights = []
convex_hull_weights = []
sum_least_squares = []
sum_convex_hull = []


def draw_vectors(mu,Sigma,n):
    return np.random.multivariate_normal(mu, Sigma, n).T  # dimxn matrix

```

```python
XX=draw_vectors(mu,Sigma,n)
```

```python
np.linalg.lstsq(XX, mu, rcond=None)
```
